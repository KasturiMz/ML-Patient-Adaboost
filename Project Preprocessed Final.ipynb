```python
#Separating the dependent and the independent variable 
x=df.drop(['PATIENT_TYPE','DEATH'],axis=1).values
y=df['PATIENT_TYPE'].values
```

```python
x
```

```python
y
```

```python
#Splitting the data into training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = 0)
```

```python
#scaling the features to obtain features on the same scale as the features may have varying magnitude, range and units.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
#Confusion metrics and accuracy score
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_pred,y_test))
```
